**Nettoyage sur les mois et les jours**
* Suppression des mois supérieur à 12
* Suppression des jours supérieurs à 31
* Quels mois on 30 jours max ? Pour supprimer les occurences > à 30 jours
Avril, Juin, Septembre, Novembre


```python
dataframeNaissance = dataframeNaissance[dataframeNaissance['month'] <= 12]
```

```python
dataframeNaissance = dataframeNaissance[dataframeNaissance['day'] <= 31]
```

```python
dataframeNaissance.drop(index = (dataframeNaissance[dataframeNaissance['month'] == 4][dataframeNaissance['day'] > 30].index), inplace=True)
dataframeNaissance.drop(index = (dataframeNaissance[dataframeNaissance['month'] == 6][dataframeNaissance['day'] > 30].index), inplace=True)
dataframeNaissance.drop(index = (dataframeNaissance[dataframeNaissance['month'] == 9][dataframeNaissance['day'] > 30].index), inplace=True)
dataframeNaissance.drop(index = (dataframeNaissance[dataframeNaissance['month'] == 11][dataframeNaissance['day'] > 30].index), inplace=True)
```

Il ne s'agit pas d'une erreur, le code s'éxécute correctement, on a juste un warning qui nous dit que la reindexation du dataframe se fait automatiquement.

**Cas du mois de Février avec les années bissextiles**

* 28 ou 29  selon l'année 
* Supprimer les individus ayant des mois supérieurs à 29 jours

```python
dataframeNaissance.drop(index = (dataframeNaissance[dataframeNaissance['month'] == 2][dataframeNaissance['day'] > 29].index), inplace=True)
```

```python
# S'intaxe pour transformer une condition en série
ma_serie = dataframeNaissance[dataframeNaissance['month'] == 2][dataframeNaissance['day'] == 29]['year']
```

iloc  (qui référence par la position) de chaque objet



```python
# Top ! La méthode isleap de calendar me renvoie True si le mois est une leap year
# Autrement dit une année bisextile - 366 jours avec 29 jours en Février
help(calendar.isleap)
```

```python
# Compréhension de liste qui me renvoie donc une LISTE
# Ici on ne récupère que les index des éléments de la serie grâce à iloc[]

# Donc on boucle autant de fois qu'il y a d'élément dans ma_serie
# Et à chaque tour on vérifie si l'année n'est pas (if not) bissextile
# Chaque index des années non bissextiles sera donc renvoyé dans la liste ma_serie.index

[ma_serie.index[i] for i in range(len(ma_serie)) if not calendar.isleap(ma_serie.iloc[i])]
```

```python
# On peut donc les supprimer grâce à une autre compréhension de liste

dataframeNaissance.drop(index = [ma_serie.index[i] for i in range(len(ma_serie)) if not calendar.isleap(ma_serie.iloc[i])] , \
inplace=True)
```
